### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
tiendas = {'tienda_1': tienda, 'tienda_2': tienda2, 'tienda_3': tienda3, 'tienda_4': tienda4}

print("Ingreso total por tienda:")
for nombre, df in tiendas.items():
    ingreso_total = df['Precio'].sum()
    print(f"  {nombre.capitalize()}: ${ingreso_total:,.2f}")
```

# 2. Ventas por categoría

```python
print("Cantidad de productos vendidos por categoría y categorías más populares por tienda:")
for nombre, df in tiendas.items():
    ventas_por_categoria = df['Categoría del Producto'].value_counts()
    print(f"  {nombre.capitalize()}:")
    print(ventas_por_categoria)
    print(f"    Categoría más popular: {ventas_por_categoria.index[0]} con {ventas_por_categoria.iloc[0]} ventas.\n")
```

# 3. Calificación promedio de la tienda


```python
print("Calificación promedio de clientes por tienda:")
for nombre, df in tiendas.items():
    calificacion_promedio = df['Calificación'].mean()
    print(f"  {nombre.capitalize()}: {calificacion_promedio:.2f}")
```

# 4. Productos más y menos vendidos

```python
print("Productos más y menos vendidos por tienda:")
for nombre, df in tiendas.items():
    conteo_productos = df['Producto'].value_counts()
    producto_mas_vendido = conteo_productos.index[0]
    cantidad_mas_vendida = conteo_productos.iloc[0]
    producto_menos_vendido = conteo_productos.index[-1]
    cantidad_menos_vendida = conteo_productos.iloc[-1]

    print(f"  {nombre.capitalize()}:")
    print(f"    Producto más vendido: {producto_mas_vendido} con {cantidad_mas_vendida} ventas.")
    print(f"    Producto menos vendido: {producto_menos_vendido} con {cantidad_menos_vendida} ventas.\n")
```

# 5. Envío promedio por tienda

```python
print("Costo de envío promedio por tienda:")
for nombre, df in tiendas.items():
    costo_envio_promedio = df['Costo de envío'].mean()
    print(f"  {nombre.capitalize()}: ${costo_envio_promedio:,.2f}")
```

# **Generando gráfico**

Genere un gráfico circular para visualizar la distribución proporcional de las categorías de productos más populares en todas las tiendas. A continuación, cree diagramas de dispersión para explorar las relaciones y los patrones entre las métricas de las tiendas, como «Ingreso Total» frente a «Calificación Promedio» o «Costo de Envío Promedio». Por último, proporcione un resumen completo de los principales hallazgos y conclusiones derivados de todos los gráficos generados.

```python
import pandas as pd

# 1. Inicializar una lista vacía para almacenar resúmenes de datos de las tiendas
datos_tiendas = []

# 2. Iterar sobre el diccionario 'tiendas'
for nombre, df in tiendas.items():
    # a. Calcular el ingreso total
    ingreso_total = df['Precio'].sum()

    # b. Calcular la calificación promedio
    calificacion_promedio = df['Calificación'].mean()

    # c. Calcular el costo de envío promedio
    costo_envio_promedio = df['Costo de envío'].mean()

    # d. Crear un diccionario con las métricas calculadas
    datos_tiendas.append({
        'Tienda': nombre.replace('_', ' ').title(),
        'Ingreso Total': ingreso_total,
        'Calificación Promedio': calificacion_promedio,
        'Costo de Envío Promedio': costo_envio_promedio
    })

# 3. Convertir la lista de diccionarios en un DataFrame
df_tiendas_resumen = pd.DataFrame(datos_tiendas)

# 4. Unir todos los DataFrames de tiendas en uno solo
df_categorias_consolidadas = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)

# 5. Calcular la cantidad de ventas por cada 'Categoría del Producto'
ventas_globales_por_categoria_counts = df_categorias_consolidadas['Categoría del Producto'].value_counts()

# 6. Convertir el resultado de value_counts() en un DataFrame y renombrar columnas
ventas_globales_por_categoria = ventas_globales_por_categoria_counts.reset_index()
ventas_globales_por_categoria.columns = ['Categoría del Producto', 'Cantidad']

print("Preparación de datos completa. 'df_tiendas_resumen' y 'ventas_globales_por_categoria' están listos.")
print("\nPrimeras filas de df_tiendas_resumen:")
print(df_tiendas_resumen.head())

print("\nPrimeras filas de ventas_globales_por_categoria:")
print(ventas_globales_por_categoria.head())
```

## Visualizar Métricas Clave de Tienda (Gráficos de Barras)

Crear gráficos de barras para comparar el ingreso total, la calificación promedio de los clientes y el costo de envío promedio entre las diferentes tiendas. Estos gráficos permitirán una comparación directa de las métricas clave de cada tienda.

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 18))

# Gráfico de Ingreso Total por Tienda
plt.subplot(3, 1, 1)
sns.barplot(x='Tienda', y='Ingreso Total', data=df_tiendas_resumen)
plt.title('Ingreso Total por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingreso Total ($)')

# Gráfico de Calificación Promedio por Tienda
plt.subplot(3, 1, 2)
sns.barplot(x='Tienda', y='Calificación Promedio', data=df_tiendas_resumen)
plt.title('Calificación Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Calificación Promedio')
plt.ylim(0, 5) # Establecer el límite del eje Y para la calificación

# Gráfico de Costo de Envío Promedio por Tienda
plt.subplot(3, 1, 3)
sns.barplot(x='Tienda', y='Costo de Envío Promedio', data=df_tiendas_resumen)
plt.title('Costo de Envío Promedio por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Costo de Envío Promedio ($)')

plt.tight_layout()
plt.show()
```

## Distribución Proporcional de Categorías de Productos (Gráfico Circular)

Generar un gráfico circular para visualizar la distribución proporcional de las categorías de productos más populares a través de todas las tiendas. Esto ayudará a identificar qué categorías dominan las ventas en términos de cantidad.

```python
plt.figure(figsize=(10, 10))
plt.pie(
    ventas_globales_por_categoria['Cantidad'],
    labels=ventas_globales_por_categoria['Categoría del Producto'],
    autopct='%1.1f%%',
    startangle=90,
    colors=sns.color_palette('viridis', len(ventas_globales_por_categoria))
)
plt.title('Distribución Proporcional de Categorías de Productos (Todas las Tiendas)')
plt.axis('equal')
plt.show()
```

## Análisis de Relaciones entre Métricas de Tienda (Gráficos de Dispersión)

Generar gráficos de dispersión para explorar las relaciones entre métricas clave de la tienda, como 'Ingreso Total' vs. 'Calificación Promedio' y 'Ingreso Total' vs. 'Costo de Envío Promedio'. Estos gráficos ayudarán a identificar patrones o correlaciones entre el rendimiento financiero y otros factores operativos.

```python
plt.figure(figsize=(12, 6))

# Gráfico de dispersión: Ingreso Total vs. Calificación Promedio
plt.subplot(1, 2, 1)
sns.scatterplot(x='Calificación Promedio', y='Ingreso Total', data=df_tiendas_resumen, hue='Tienda', s=100)
plt.title('Ingreso Total vs. Calificación Promedio')
plt.xlabel('Calificación Promedio')
plt.ylabel('Ingreso Total ($)')
plt.grid(True)

# Gráfico de dispersión: Ingreso Total vs. Costo de Envío Promedio
plt.subplot(1, 2, 2)
sns.scatterplot(x='Costo de Envío Promedio', y='Ingreso Total', data=df_tiendas_resumen, hue='Tienda', s=100)
plt.title('Ingreso Total vs. Costo de Envío Promedio')
plt.xlabel('Costo de Envío Promedio ($)')
plt.ylabel('Ingreso Total ($)')
plt.grid(True)

plt.tight_layout()
plt.show()
```
